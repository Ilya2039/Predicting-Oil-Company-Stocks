```python
# Отчетность по акциям OXY

# таблица с EPS и доходом и с их прогнозами 
data = pd.read_excel('OXY_Доход.xlsx')

# Меняем имена столбцов для лучшего восприятия
#data.rename(columns={'Прогноз': 'Прогноз EPS'}, inplace=True)
#data.rename(columns={'Прогноз.1': 'Прогноз Дохода'}, inplace=True)

# Приводим элементы в таблице в типу floar(В 2 прогнозах)
def stringCleaner(s): 
    # initialize an empty list
    str1 = [] 
    chars = set('0123456789,+-MB')
    for c in s:
        if c in chars:
            str1.append(c) 
    return ''.join(str1)

for i in range(len(data['Прогноз EPS'])):
    data['Прогноз EPS'][i] = (stringCleaner(str(data['Прогноз EPS'][i])))

for i in range(len(data['Прогноз Дохода'])):
    data['Прогноз Дохода'][i] = (stringCleaner(str(data['Прогноз Дохода'][i])))


# Убираем последние строчки и первую
data = data.head(33)
data = data[1:]

# Выводим изменную таблицу
#data = data.iloc[::-1]
data
```

```python
XOM = yf.Ticker("OXY")

# get stock info
#msft.info

# get historical market data
price = XOM.history(period="max")

# Достаем 2000 последних дней
price = price.tail(1976)

# Добавляем индексы слева
price = price.reset_index()

# Создаем колонку где будет макс цена акции за день
price['max_per_day'] = 0.0

# Добавляем в колонку max_per_day максимальную цену акции за день
for i in range(len(price['max_per_day'])):
    price['max_per_day'][i] = max(price['Open'][i], price['High'][i], price['Low'][i], price['Close'][i])

# Выводим таблицу после добавления столбца max_per_day
price
```

```python
# Добавляем в таблицу data колонку max_price для подсчета максимальной цены в следубщем квартале после того как вышел квартальный отчет
data['max_price'] = 0.0

# считаем макс цену акции за следующий квартал после объявления отчета
for i in range(len(data['Дата выпуска']), 0, -1):
    for j in range(len(price['Date'])):      
        if data['Дата выпуска'][i] == price['Date'][j]:
            max0 = 0
            if i > 1:
                while price['Date'][j] != data['Дата выпуска'][i - 1]:
                    if price['max_per_day'][j] > max0:
                        max0 = price['max_per_day'][j]
                    j += 1
            if i == 1:
                while j <= 1974:
                    if price['max_per_day'][j] > max0:
                        max0 = price['max_per_day'][j]
                    j += 1
                    
            
            data['max_price'][i] = max0
            
data
```

```python
# Переделываем столбцы Доход и Прогноз Дохода в формат float избавляясь от B и M и меняем ',' на '.' для перевода из str во float
for i in range(1, len(data['Доход']) + 1):
    if str(data['Доход'][i][-1]) == 'B':
        data['Доход'][i] = (data['Доход'][i][0:len(data['Доход'][i]) - 1])
        data['Доход'][i] = str(data['Доход'][i]).replace(',', '.')
        data['Доход'][i] = float(data['Доход'][i])
        data['Доход'][i] = data['Доход'][i] * 1000000000
    elif str(data['Доход'][i][-1]) == 'M':
        data['Доход'][i] = (data['Доход'][i][0:len(data['Доход'][i]) - 1])
        data['Доход'][i] = str(data['Доход'][i]).replace(',', '.')
        data['Доход'][i] = float(data['Доход'][i])
        data['Доход'][i] = data['Доход'][i] * 1000000000
        
for i in range(1, len(data['Прогноз Дохода']) + 1):
    if str(data['Прогноз Дохода'][i][-1]) == 'B':
        data['Прогноз Дохода'][i] = (data['Прогноз Дохода'][i][0:len(data['Прогноз Дохода'][i]) - 1])
        data['Прогноз Дохода'][i] = str(data['Прогноз Дохода'][i]).replace(',', '.')
        data['Прогноз Дохода'][i] = float(data['Прогноз Дохода'][i])
        data['Прогноз Дохода'][i] = data['Прогноз Дохода'][i] * 1000000000
    elif str(data['Прогноз Дохода'][i][-1]) == 'M':
        data['Прогноз Дохода'][i] = (data['Прогноз Дохода'][i][0:len(data['Прогноз Дохода'][i]) - 1])
        data['Прогноз Дохода'][i] = str(data['Прогноз Дохода'][i]).replace(',', '.')
        data['Прогноз Дохода'][i] = float(data['Прогноз Дохода'][i])
        data['Прогноз Дохода'][i] = data['Прогноз Дохода'][i] * 1000000
        
for i in range(1, len(data['Прогноз EPS']) + 1):
    data['Прогноз EPS'][i] = str(data['Прогноз EPS'][i]).replace(',', '.')
    data['Прогноз EPS'][i] = float(data['Прогноз EPS'][i])
    
data
```
